**Image to Pencil Sketch with Python**

**DESCRIPTION TO TASK** : We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python

**IMPORTING LIBRARIES**

```python
from matplotlib import  pyplot as plt
import cv2
```

**READ THE IMAGE**

```python
image1 = cv2.imread("cheetah.jpeg")
plt.figure(figsize=(8,8))
plt.imshow(image1)
plt.show()
```

**CONVERTING THE IMAGE TO RGB COLORS**

```python
rgbimage=cv2.cvtColor(image1,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(rgbimage)
plt.show()
```

**CONVERTING THE RGB IMAGE INTO A GRAY SCALE IMAGE**

```python
grayimage1=cv2.cvtColor(rgbimage, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(grayimage1,cmap='gray')
plt.show()
```

**CONVERTING THE GRAYSCALE IMAGE INTO A INVERTED IMAGE**

```python
invertgrayimage1 = 255 - grayimage1
plt.figure(figsize=(8,8))
plt.imshow(invertgrayimage1,cmap='gray')
plt.show()
```

**BLUR THE INVERTED GRAYSCALE IMAGE**

```python
blurimg = cv2.GaussianBlur(invertgrayimage1,(21,21),0)
plt.figure(figsize=(8,8))
plt.imshow(blurimg,cmap='gray')
plt.show()
```

**INVERT THE BLURRED IMAGE**

```python
invert_blurimg=255 - blurimg
plt.figure(figsize=(8,8))
plt.imshow(invert_blurimg,cmap='gray')
plt.show()
```

**DIVIDE THE GRAYSCALE IMAGE WITH INVERTED BLURRED IMAGE TO GET THE SKETCH**

```python
sketch=cv2.divide(grayimage1,invert_blurimg,scale=256.0)
plt.figure(figsize=(8,8))
plt.imshow(rgbimage)
plt.figure(figsize=(8,8))
plt.imshow(sketch,cmap='gray')
plt.show()
```
