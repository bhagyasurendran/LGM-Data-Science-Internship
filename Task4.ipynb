**LABEL ENCODER**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label=LabelEncoder()
df1=df.copy()
df1['Species']=label.fit_transform(df1['Species'])
df1
```

```python
x=df1.drop(columns=['Species'],axis=1)
y=df1['Species']
```

**TRAINING AND TESTING**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.6)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
print(y_predict[0:10])
```

```python
sns.heatmap(confusion_matrix(y_test, y_predict),annot=True)
```

```python
accuracy = accuracy_score(y_test,y_predict)
print("Accuracy Score is:",accuracy)
```

**CLASSIFICATION REPORT**

```python
print(classification_report(y_test, y_predict))
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(16,10))
plot_tree(model,feature_names=df1.columns[:-1],class_names=df.columns[-1],rounded=True,filled=True)
```

**EVALUATION OF  MODEL**

```python
train_score= str(model.score(x_train, y_train) * 100)
test_score= str(model.score(x_test, y_test) * 100)
print(f'Train Score : {train_score[:5]}%\nTest Score : {test_score[:5]}%')
```
